```python
sns.boxplot(df["concavity_worst"])
```

```python
for x in df.columns[::]:
    if df[x].dtype==object:
        continue
    else:
        q1=df[x].quantile(q=0.25)
        q3=df[x].quantile(q=0.75)
        IQR= q3-q1
        
        low=q1-1.5*IQR
        high=q3+1.5*IQR
        df.loc[(df[x]<low)| (df[x]>high),x]=df[x].median()
```

```python
sns.boxplot(df['concavity_worst'])
```

```python
df=pd.get_dummies(df,drop_first=True)
print(df.shape)
```

```python
df.corr()
```

```python
df.isnull().sum()
```

```python
x =df.drop('diagnosis_M',axis=1)
y = df["diagnosis_M"].values.reshape(-1,1)
```

```python
# standardization
from sklearn.preprocessing import MinMaxScaler
# define Min max  Scaler 
mn = MinMaxScaler()
# transform data
scaled = mn.fit_transform(x)
```

```python
# import the model
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(scaled,y, test_size=0.2,random_state=40,stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# instantiate 
model = LogisticRegression()
```

```python
#  fit the model with data 
model.fit(x_train,y_train)
```

```python
# predict wiht test data 
y_pred = model.predict(x_test)
y_train_pred = model.predict(x_train)
```

```python
model.score(x_train,y_train)
```

```python
# import the matrics class  for test
from sklearn.metrics import confusion_matrix 
cnf_matrix = confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
# import the matrics class  for train 
from sklearn.metrics import confusion_matrix 
cnf = confusion_matrix(y_train, y_train_pred)
cnf
```

```python
### Report for Test
from sklearn.metrics import classification_report 

print(classification_report(y_test,y_pred))
```

```python
##### report for y_train 
from sklearn.metrics import classification_report 

print(classification_report(y_train,y_train_pred))
```

```python
# predict performance measure for y_test
from sklearn import metrics 

print("Accuracy = ",metrics.accuracy_score(y_test,y_pred))
print("Precision = ",metrics.precision_score(y_test,y_pred))
print("Recall = ",metrics.recall_score(y_test,y_pred))
```

```python
# predict performance measure for y_train 
from sklearn import metrics 

print("Accuracy = ",metrics.accuracy_score(y_train,y_train_pred))
print("Precision = ",metrics.precision_score(y_train,y_train_pred))
print("Recall = ",metrics.recall_score(y_train,y_train_pred))
```
